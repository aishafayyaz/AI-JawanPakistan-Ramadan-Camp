```python
print("My Calculator")
x= int(input("Enter 1st number:"))
operand = input("Enter operation to be performed:")
y= int(input("Enter 2nd number:"))
if (operand == "+"):
    print(x+y)
elif (operand == "-"):
    print(x-y)
elif (operand=="*"):
    print(x*y)
elif (operand=="/"):
    if (y==0):
        print ("Divisor can not be zero!")
    else:
        print(x/y)
else:
    print("Invalid operator")
```
