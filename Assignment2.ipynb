```python
# take input from user, converts input to list and prints length of the list
data=input("Enter elements of list separated by space: ")
user_list = data.split()
print(user_list)
print("The length of your entered list is : ", len(user_list))
```

```python
# calculate sum of numbers of the list
my_input=input("Enter numbers of your list separated by space :") 
my_list=list(map(int,my_input.split())) 
print(my_list) 
print("The sum of numbers in the list is: ", sum(my_list))
```

```python
# get the largest number from the list
largest=max(my_list)
print("The largest number in your list is ", largest)
```

```python
# lists all numbers of a list which are less than 5
my_input=input("Enter numbers of your list separated by space :") 
my_list=list(map(int,my_input.split())) 
print(my_list) 
for x in my_list:
    if x < 5:
        print(x)
```
