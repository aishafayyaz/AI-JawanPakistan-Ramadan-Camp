```python
print("\t\t....MArkSheet Generator.....\n")
Name=input("Enter Student name:")
Math=float(input("Enter Math Marks : "))
English=float(input("Enter English Marks : "))
Urdu=float(input("Enter Urdu Marks : "))
Science=float(input("Enter Science Marks : "))
Computer=float(input("Enter Computer Marks : "))
Obt_Marks = Math + English + Urdu + Science + Computer
Percentage = (Obt_Marks/500) * 100
if (Percentage <=100 and Percentage >= 95):
    grade = "A++"
    remarks = "Exceptional"
elif (Percentage <95 and Percentage >= 90):
    grade = "A+"
    remarks = "Outstanding"
elif (Percentage <90 and Percentage >= 85):
    grade = "A"
    remarks = "Excellent"
elif (Percentage <85 and Percentage >= 80):
    grade = "B++"
    remarks = "Very Good"
elif (Percentage <80 and Percentage >= 75):
    grade = "B+"
    remarks = "Good"
elif (Percentage <75 and Percentage >= 70):
    grade = "B"
    remarks = "Fairly Good"
elif (Percentage <70 and Percentage >= 60):
    grade = "C" 
    remarks = "Above Average"
elif (Percentage <60 and Percentage >= 50):
    grade = "D" 
    remarks = "Average"
elif (Percentage <50 and Percentage >= 40):
    grade = "E" 
    remarks = "Below Average"
else:
    grade = "U"
    remarks = "Unsatisfactory"
print("\n\t\t........Result........\n")
print("Name: ",Name)
print("Obtained Marks: ", Obt_Marks)
print("Percentage: ",Percentage,"%")
print("Grade: ", grade)
print("Remarks: ",remarks)
```
